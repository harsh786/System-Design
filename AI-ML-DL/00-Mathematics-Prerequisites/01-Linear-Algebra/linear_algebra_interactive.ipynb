## 1. Vector Operations

Vectors are the building blocks of linear algebra. In ML, data points are vectors in high-dimensional space.

```python
# Define vectors
a = np.array([3, 2])
b = np.array([1, 4])

# Basic operations
print(f"a = {a}")
print(f"b = {b}")
print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"2 * a = {2 * a}")
print(f"Dot product a·b = {np.dot(a, b)}")
print(f"||a|| (magnitude) = {np.linalg.norm(a):.4f}")
print(f"||b|| (magnitude) = {np.linalg.norm(b):.4f}")
```

```python
# Visualize vectors
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Plot vectors
for ax, title, vecs, colors in zip(
    axes,
    ["Vectors a and b", "Vector Addition", "Vector Subtraction"],
    [[a, b], [a, b, a+b], [a, b, a-b]],
    [["blue", "red"], ["blue", "red", "green"], ["blue", "red", "purple"]]
):
    ax.set_xlim(-1, 6)
    ax.set_ylim(-3, 7)
    ax.set_aspect("equal")
    ax.axhline(y=0, color="k", linewidth=0.5)
    ax.axvline(x=0, color="k", linewidth=0.5)
    ax.set_title(title)
    labels = ["a", "b", "a+b" if "Add" in title else "a-b"]
    for v, c, l in zip(vecs, colors, labels):
        ax.arrow(0, 0, v[0], v[1], head_width=0.15, head_length=0.1, fc=c, ec=c)
        ax.annotate(l, xy=v, fontsize=12, color=c)

plt.tight_layout()
plt.show()
```

### Dot Product and Cosine Similarity

The dot product measures how "aligned" two vectors are:

```python
# Cosine similarity
cos_sim = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
angle = np.arccos(cos_sim)
print(f"Cosine similarity: {cos_sim:.4f}")
print(f"Angle between vectors: {np.degrees(angle):.2f} degrees")

# Orthogonal vectors have cosine similarity = 0
c = np.array([2, -3])
d = np.array([3, 2])
print(f"
c = {c}, d = {d}")
print(f"c·d = {np.dot(c, d)} (orthogonal!)")
```

## 2. Matrix Operations

Matrices represent linear transformations. In ML:
- Weight matrices transform input features
- Data matrices hold multiple samples as rows

```python
# Matrix creation
A = np.array([[1, 2], [3, 4], [5, 6]])
B = np.array([[7, 8, 9], [10, 11, 12]])

print(f"A (3x2):
{A}
")
print(f"B (2x3):
{B}
")
print(f"A @ B (3x3):
{A @ B}
")
print(f"Shape of A @ B: {(A @ B).shape}")
```

```python
# Visualize matrix as linear transformation
T = np.array([[2, 1], [0, 1.5]])  # Transformation matrix

# Create a unit square
square = np.array([[0,0],[1,0],[1,1],[0,1],[0,0]]).T
transformed = T @ square

fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.plot(*square, "b-o", label="Original")
ax.plot(*transformed, "r-o", label="Transformed")
ax.set_xlim(-0.5, 4)
ax.set_ylim(-0.5, 3)
ax.set_aspect("equal")
ax.legend()
ax.set_title(f"Linear Transformation by T = {T.tolist()}")
ax.grid(True)
plt.show()
```

## 3. Eigenvalues and Eigenvectors

For a square matrix $, if \mathbf{v} = \lambda\mathbf{v}$, then:
- $\mathbf{v}$ is an eigenvector (direction unchanged by transformation)
- $\lambda$ is the eigenvalue (scaling factor)

**Why it matters in ML:** PCA uses eigenvectors to find principal directions of variance.

```python
# Compute eigenvalues and eigenvectors
M = np.array([[4, 2], [1, 3]])
eigenvalues, eigenvectors = np.linalg.eig(M)

print(f"Matrix M:
{M}
")
print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):
{eigenvectors}
")

# Verify: M @ v = lambda * v
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]
    print(f"M @ v{i+1} = {M @ v}")
    print(f"λ{i+1} * v{i+1} = {lam * v}")
    print()
```

```python
# Visualize eigenvectors
fig, ax = plt.subplots(figsize=(6, 6))

# Plot random vectors and their transformations
np.random.seed(42)
for _ in range(20):
    v = np.random.randn(2)
    v = v / np.linalg.norm(v)
    Mv = M @ v
    ax.arrow(0, 0, v[0], v[1], head_width=0.05, color="lightblue", alpha=0.5)
    ax.arrow(0, 0, Mv[0], Mv[1], head_width=0.05, color="lightcoral", alpha=0.5)

# Plot eigenvectors (they only get scaled, not rotated)
for i, color in enumerate(["blue", "red"]):
    v = eigenvectors[:, i]
    ax.arrow(0, 0, v[0], v[1], head_width=0.08, fc=color, ec=color, linewidth=2)
    Mv = M @ v
    ax.arrow(0, 0, Mv[0], Mv[1], head_width=0.08, fc=color, ec=color, linewidth=2, linestyle="dashed")
    ax.annotate(f"eigvec {i+1} (λ={eigenvalues[i]:.1f})", xy=v*1.1, color=color, fontsize=10)

ax.set_xlim(-3, 5)
ax.set_ylim(-2, 4)
ax.set_aspect("equal")
ax.set_title("Eigenvectors remain in same direction after transformation")
plt.show()
```

## 4. Principal Component Analysis (PCA)

PCA finds the directions of maximum variance in data:
1. Center the data (subtract mean)
2. Compute covariance matrix
3. Find eigenvectors of covariance matrix
4. Project data onto top-k eigenvectors

```python
# Generate correlated 2D data
np.random.seed(42)
mean = [2, 3]
cov = [[3, 2], [2, 2]]  # Correlated
data = np.random.multivariate_normal(mean, cov, 200)

# Step 1: Center data
data_centered = data - data.mean(axis=0)

# Step 2: Covariance matrix
cov_matrix = np.cov(data_centered.T)
print(f"Covariance matrix:
{cov_matrix}
")

# Step 3: Eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
# Sort by descending eigenvalue
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

print(f"Eigenvalues: {eigenvalues}")
print(f"Variance explained: {eigenvalues / eigenvalues.sum() * 100}%")
```

```python
# Visualize PCA
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original data with principal components
ax = axes[0]
ax.scatter(data_centered[:, 0], data_centered[:, 1], alpha=0.5, s=20)
for i, (ev, evec) in enumerate(zip(eigenvalues, eigenvectors.T)):
    ax.arrow(0, 0, evec[0]*ev, evec[1]*ev, head_width=0.1, fc=["red","blue"][i], ec=["red","blue"][i], linewidth=2)
    ax.annotate(f"PC{i+1} (var={ev:.2f})", xy=evec*ev*1.1, fontsize=10, color=["red","blue"][i])
ax.set_aspect("equal")
ax.set_title("Data with Principal Components")
ax.set_xlabel("x1"); ax.set_ylabel("x2")

# Project onto PC1 (dimensionality reduction 2D -> 1D)
projected = data_centered @ eigenvectors[:, 0:1]
reconstructed = projected @ eigenvectors[:, 0:1].T

ax = axes[1]
ax.scatter(data_centered[:, 0], data_centered[:, 1], alpha=0.2, s=20, label="Original")
ax.scatter(reconstructed[:, 0], reconstructed[:, 1], alpha=0.5, s=20, c="red", label="Projected to PC1")
ax.set_aspect("equal")
ax.set_title("PCA: 2D → 1D Projection")
ax.legend()

plt.tight_layout()
plt.show()
```
